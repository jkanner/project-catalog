### This file saves hdf5 file catalog entries as feather files in a posterior_chains folder

```python
import pandas as pd
```

```python
def get_datasets(file_path):
    store = pd.HDFStore(file_path, mode='r')
    datasets = store.keys()
    store.close()
    return datasets

def save_files(file_path):
    datasets = get_datasets(file_path)
    for dataset in datasets:
        name = 'LDC2A_' + dataset.split('_')[-2]
        df = pd.read_hdf(file_path, dataset)
        df.to_feather(f'/Users/aaron/Documents/lisa/project-catalog/data/erebor_posterior_chains/{name}.feather')
```

```python
for i in range(10, 24):
    save_files(f'/Users/aaron/Documents/lisa/project-catalog/data/erebor_training_catalogs_v1.0.4/erebor_training_catalogs_v1.0.4_gb_all_chain_file_00{i}.h5')
```
