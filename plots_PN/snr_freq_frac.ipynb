```python
from project_catalog.utils import get_galactic_binary_names
from project_catalog.galactic_binary import GalacticBinary
from project_catalog import compute_1d_hdrs, compute_sky_location_hdr, compute_8d_hdr

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm
import healpy as hp
from astropy.coordinates import SkyCoord
import astropy.units as u

# Load injection names and HDR data
names = get_galactic_binary_names()
hdr_df = pd.read_feather("../data/hdrs.feather")

data = []

for name in tqdm(names):
    gb = GalacticBinary.load_feather(name)
    if gb.candidates is None or gb.candidates.empty:
        continue

    hdr_rows = hdr_df[hdr_df['Name'] == name]
    if hdr_rows.empty:
        continue  # no acceptable candidate

    best_row = hdr_rows.loc[hdr_rows['8D HDR'].idxmin()]
    best_id = best_row['Candidate']
    candidate_row = gb.candidates[gb.candidates['Name'] == best_id]

    inj_freq = candidate_row['Frequency'].values[0]
    snr = candidate_row['SNR'].values[0]
    freq_hdr = best_row['Frequency HDR']
    in_hdr = freq_hdr <= 0.9

    data.append({
        'frequency': inj_freq,
        'log_snr': np.log10(snr),
        'in_hdr': in_hdr
    })

# Create DataFrame
df = pd.DataFrame(data).dropna(subset=['frequency', 'log_snr', 'in_hdr'])

# Define bin edges (half the number of bins → double bin size)
num_bins = 50  # was 100 before
freq_bins = np.linspace(df['frequency'].min(), df['frequency'].max(), num_bins + 1)
snr_bins = np.linspace(df['log_snr'].min(), df['log_snr'].max(), num_bins + 1)

# Compute bin centers for plotting
freq_centers = 0.5 * (freq_bins[:-1] + freq_bins[1:])
snr_centers = 0.5 * (snr_bins[:-1] + snr_bins[1:])

# Assign bin indices
df['freq_bin'] = pd.cut(df['frequency'], bins=freq_bins, labels=False, include_lowest=True)
df['snr_bin'] = pd.cut(df['log_snr'], bins=snr_bins, labels=False, include_lowest=True)

# Create 2D array for HDR recovery fraction
heatmap_data = np.full((num_bins, num_bins), np.nan)

for i in range(num_bins):
    for j in range(num_bins):
        mask = (df['freq_bin'] == i) & (df['snr_bin'] == j)
        if mask.any():
            heatmap_data[j, i] = df.loc[mask, 'in_hdr'].mean()

# Plotting
X, Y = np.meshgrid(freq_centers, snr_centers)

plt.figure(figsize=(10, 6))
# Force color scale from 0.5 to 1
heatmap = plt.pcolormesh(X, Y, heatmap_data, cmap='coolwarm', shading='auto', vmin=0.5, vmax=1.0)
cbar = plt.colorbar(heatmap, label='Fraction in 90% HDR')
cbar.set_ticks(np.linspace(0.5, 1.0, 6))

plt.ylabel("log10(SNR)")
plt.xlabel("Injection Frequency [Hz]")
plt.title("Fraction of Sources Recovered within 90% HDR\nvs Frequency and SNR")
plt.grid(False)
plt.tight_layout()
plt.savefig("fraction_recovered_vs_freq_log_snr_half_colorbar.pdf")
plt.show()
```
